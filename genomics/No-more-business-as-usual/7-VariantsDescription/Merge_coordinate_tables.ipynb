# Merge maps file into a single coordinate conversion table

Inputs : Files .map generated by MAFFT --add

Upload all tables as dataframes

```python
fileslist=glob.glob("*.fasta.map")
dic={}
for file in fileslist : 
    coord=pd.read_csv(file,sep=',', header=0, skiprows=1,na_values=0)
    name=file.split('.')[0]
    coord.columns=['AA',name,'Alignment']
    dic[name]=coord
```

Join tables on alignement coordinate to the refererence genome (here NC_045512 )

```python
res=dic['NC_045512']
for i in dic.keys():
    if i!='NC_045512' :
        tmp=dic[i][[i,'Alignment']]
        if np.array_equal(dic[i].Alignment, dic[i].Alignment.astype(str)):
            tmp['Alignment']=[int(x.strip().replace('-','0')) for x in tmp['Alignment']]
        res=pd.merge_ordered(res, tmp, fill_method=None, left_by='Alignment')
        res[i]=res[i].astype('Int64')
```

Replace NA by zeros and write the resulting table in a tablullated file

```python
res=res.replace(np.nan,0)
res.to_csv(repos+'/coordinates.tab',  sep='\t',index=False, header=True)
```
